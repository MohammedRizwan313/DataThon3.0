```python
html_data = requests.get("https://www.mfa.gov.cn/")
clean_data = BeautifulSoup(html_data.content, "html.parser")
chinese_text = clean_data.find_all(text=True)

translator = Translator(to_lang="en")

for idx, text in enumerate(chinese_text):
    if len(text) > 0 and any('\u4e00' <= char <= '\u9fff' for char in text):
        english_translation = translator.translate(text)
        chinese_text[idx] = english_translation
print(" ".join(chinese_text))
```
